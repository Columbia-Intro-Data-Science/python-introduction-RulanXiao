```python
# read data into a DataFrame
import numpy
import pandas as pd
import pylab as plt
import seaborn
from sklearn.linear_model import LinearRegression
import numpy.random as nprnd
import random
import json
pd.set_option('display.max_columns', 500)
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

df = pd.read_csv('/Users/rulanxiao/Desktop/Book1.csv')
df.head()
```

```python
df.corr()
```

```python
y = df['sales']
X = df[['prom','adv']]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=48)

regr = LinearRegression() # Create linear regression object
regr.fit(X_train, y_train) # Train the model using the training sets

plt.title('R^2 on testing data')
x=regr.predict(X_test)  #predicted value
y=y_test   # actual value
plt.scatter(x,y)
plt.plot(y,y)

print('Residual sum of squares: %.2f'  #R^2 = residual sum of square
      % numpy.mean((regr.predict(X_test) - y_test) ** 2))
print('R^2 Score: %.2f' % regr.score(X_test, y_test))

## A small R^2 would indicate a tight fit of the model to the data
## For this testing data, R^2 score is good meaning predicted and actual very close.
```
