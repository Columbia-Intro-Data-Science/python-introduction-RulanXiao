## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations,lrate):  ## add variable "lrate" of learning rate to function
    points = genfromtxt("/Users/rulanxiao/Documents/GitHub/APMAE4990-/data/data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    learning_rate = lrate
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1],color='b')
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100,0.0001)
````

````python
# Use this to visually discuss convergence rate based on learning rate

##Here the well-working learning rate is 0.0001
learning_rate = 0.0001
for num in range(0,10):
    run(num,learning_rate)
    plt.show()
````

````python
## Try to change learning rate smaller to 0.00001
learning_rate = 0.00001
for num in range(0,10):
    run(num,learning_rate)
    plt.show()
````

````python
## Try to change learning rate bigger to 0.0005
learning_rate = 0.0005
for num in range(0,10):
    run(num,learning_rate)
    plt.show()
````

````python
## Try to change learning rate much bigger to 0.01
learning_rate = 0.01
for num in range(0,10):
    run(num,learning_rate)
    plt.show()
````

````python
## Try to change learning rate much bigger to 1
learning_rate = 1
for num in range(0,10):
    run(num,learning_rate)
    plt.show()
````

For learning rate = 0.00001, it has a ver slow convergence. 

For learning rate = 0.0001, it has a fast convergence.

For learning rate = 0.0005, it has no convergence

For learning rate = 0.001, no convergence.

For learning rate = 1, no convergence.

We may say learning rate = 0.0001 is the optimal choice. 

We are trying to use gradient descent method find $\beta_*$ that will minimize $\mathcal{L}$. 

$\beta_{n} = \beta_{n-1} - \kappa \nabla_{\beta} \mathcal{L}(\beta_{n-1})$.
(in our code like "new_b=b_current-(learningRate*b_gradient)

Then the learning rate determines the time step sizes that gradient descent use to find $\beta_*$, which lead to the "lowest" point of the convex function. 

Then if the time step size (learning rate) too small, it will make $\beta_{n+1}$and $\beta_n$ very close. And that will cause a large number of iterations. That is, smaller learning size will cause larger number of iterations.

But it doesn't means learning rate to be larger the better. We would like to make a careful choice on learning rate, because we may easily miss the rate that reaches $\beta_*$ if we take large steps. If so, it will be divergent at the end.

````python
# we choose learning rate =0.0001 here:
def runerr(num_iterations,learning_rate):
    points = genfromtxt("/Users/rulanxiao/Documents/GitHub/APMAE4990-/data/data.csv", delimiter=",")
    initial_b = 0 
    initial_m = 0 
    num_iterations = num_iterations
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    err = compute_error_for_line_given_points(b, m, points)
    return err

error1 = np.empty(100)
error2 = np.empty(100)
error3 = np.empty(100)
error4 = np.empty(100)
error5 = np.empty(100)
number=range(0, 100)

for m, i in enumerate(range(0, 100)):
    error1[m] = runerr(i,0.00001)
    
for m, i in enumerate(range(0, 100)):
    error2[m] = runerr(i,0.0001)
    
for m, i in enumerate(range(0, 100)):
    error3[m] = runerr(i,0.0005)

for m, i in enumerate(range(0, 100)):
    error4[m] = runerr(i,0.001)
    
for m, i in enumerate(range(0, 100)):
    error5[m] = runerr(i,1)

plt.plot(number, error1, 'b')
plt.xlabel('Number of iterations')
plt.ylabel('Error for LR=0.00001')
plt.title('Error vs Number of iterations, LR=0.00001')
plt.show()

plt.plot(number, error2, 'b')
plt.xlabel('Number of iterations')
plt.ylabel('Error for LR=0.0001')
plt.title('Error vs Number of iterations, LR=0.0001')
plt.show()

plt.plot(number, error3, 'b')
plt.xlabel('Number of iterations')
plt.ylabel('Error for LR=0.0005')
plt.title('Error vs Number of iterations, LR=0.0005')
plt.show()

plt.plot(number, error4, 'b')
plt.xlabel('Number of iterations')
plt.ylabel('Error for LR=0.001')
plt.title('Error vs Number of iterations, LR=0.001')
plt.show()

plt.plot(number, error5, 'b')
plt.xlabel('Number of iterations')
plt.ylabel('Error for LR=1')
plt.title('Error vs Number of iterations, LR=1')
plt.show()

## we see when learning rate= 0.0001, we only need number of iteration less than 10.
````
